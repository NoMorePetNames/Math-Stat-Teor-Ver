### 1.
Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].  
Найдите ее среднее значение и дисперсию.

Среднее значение:  
$$M(X) = \frac{a + b}{2} = \frac{200 + 800}{2} = 500$$

Дисперсия:  
$$D(X) = \frac{(b - a)^2}{12} = \frac{(800 - 200)^2}{12} = 30000$$

### 2. 
О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.  
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?  
Если да, найдите ее.  

Выразим правую границу из формулы дисперсии для непрерывной равномерно распределенной величины:

$$D(X) = \frac{(b - a)^2}{12}$$

$$b^2 - 2ab + a^2 = 12D$$

При D = 0.2 и a = 0.5 получаем:

$$b^2 - b + 0.25 = 2.4$$

$$b^2 - b - 2.15 = 0$$

$$b_1 = \frac{1 - \sqrt{(-1)^2 - 4\cdot1\cdot(-2.15)}}{2\cdot1}= -1.049$$

$$b_2 = \frac{1 + \sqrt{(-1)^2 - 4\cdot1\cdot(-2.15)}}{2\cdot1}= 2.049$$

Учитывая, что правая граница должна находится правее левой на числовой оси, значение правой границы:

$$b = b_2 = 2.049$$

Тогда среднее значение:

$$M(X) = \frac{a + b}{2} = \frac{0.2 + 2.049}{2} = 1.1245$$

### 3.  
Непрерывная случайная величина X распределена нормально и задана плотностью распределения

$$f(x) = \frac{1}{4 \sqrt{2 \pi}} e ^ {- \frac{(x + 2)^2}{32}}$$

Найдите:  
а).M(X)  
б).D(X)  
в).std(X)(среднее квадратичное отклонение)

Из формулы плотности вероятности для нормального распределения:

$$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}} = \frac{1}{4 \sqrt{2 \pi}} e ^ {- \frac{(x + 2)^2}{32}},$$
где:

$\sigma^2 = D(x) = 16$   

$\sigma = std(x) = 4$ 

$a = M(x) = -2$

### 5.     
На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см,  
от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

$\sigma^2 = D(x) = 25$   

$\sigma = std(x) = 5$   

M(x) = 178$

$x = 190$

$\frac{M(x) - x}{\sigma} = \frac{178 - 190}{5} = - \frac{12}{5}$

#### Значение 190см отклоняется от математического ожидания 178см на $2\frac{2}{5}$ стандартных отклонений при D(X) = 25

```python
#Код ниже строит график нормального распределения с задаными характеристиками
import random
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normal_distribution(x, mu, sigma):
    return 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2))
```

```python
def plot_distribution(mu, sigma, a=None, b=None, a2=None, b2=None):
    """Строит нормальное распределение,
    если необходимо то закрашивает участок под кривой [a, b]
    my - математическое ожидание
    sigma - среднеквадратичное отклонение
    [a, b] - участок под кривой, площадь которого необходимо найти"""
    
    #координаты точек для построения кривой нормального распределения
    x = [i for i in range(mu - 4*sigma, mu + 4*sigma)]
    y = [normal_distribution(j, mu, sigma) for j in x]
    
    #координаты точек для прямой x = my
    x_mu = [mu]*2
    y_mu = min(y), max(y)
    
    #координаты для линий, которые ограничивают площадь под кривой [a, b]
    if a and b:
        x_ab = [i for i in range(a, b+1)]
        y_ab = [normal_distribution(j, mu, sigma) for j in x_ab]
        #Дублируем крайние координаты по оси х, чтобы ограничить штрихуемую площадь вертикальными прямыми
        x_ab = x_ab[:1] + x_ab + x_ab[-1:]
        y_ab = [min(y),] + y_ab + [min(y),]
        
        #Координаты вертикальных прямых x=a и x=b
        x_a = [a]*2
        y_a = [min(y), normal_distribution(a, mu, sigma)]
        
        x_b = [b]*2
        y_b = [min(y), normal_distribution(b, mu, sigma)]
    
    if a2 and b2:
        x_ab2 = [i for i in range(a2, b2+1)]
        y_ab2 = [normal_distribution(j, mu, sigma) for j in x_ab2]
        #Дублируем крайние координаты по оси х, чтобы ограничить штрихуемую площадь вертикальными прямыми
        x_ab2 = x_ab2[:1] + x_ab2 + x_ab2[-1:]
        y_ab2 = [min(y),] + y_ab2 + [min(y),]
        
        #Координаты вертикальных прямых x=a и x=b
        x_a2 = [a2]*2
        y_a2 = [min(y), normal_distribution(a2, mu, sigma)]
        
        x_b2 = [b]*2
        y_b2 = [min(y), normal_distribution(b2, mu, sigma)]
        
    plt.figure(figsize =(6, 4))
    plt.plot(x, y, color = 'black')
    plt.plot(x_mu, y_mu, color='black', linestyle = 'dashdot', label='M(X)')
    plt.title('Распределение роста взрослого населения города Х')
    plt.xlabel('Рост')
    plt.ylabel('Плотность вероятности')
    
    plt.grid()
    if a and b:
        plt.fill(x_ab, y_ab, 'lightgray')
        plt.plot(x_a, y_a, color='black', linestyle = 'dashed', label='Границы а и b')
        plt.plot(x_b, y_b, color='black', linestyle = 'dashed')
    if a2 and b2:
        plt.fill(x_ab2, y_ab2, 'lightgray')
        plt.plot(x_a2, y_a2, color='black', linestyle = 'dashed')
        plt.plot(x_b2, y_b2, color='black', linestyle = 'dashed')
    plt.legend()
    plt.show()
```
